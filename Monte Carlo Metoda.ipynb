```python
import numpy as np
from matplotlib import pyplot as plt
```

### Izracunavanje broja Pi pomocu Monte Karlo metode. 

```python
N = np.arange(1000,100000,1000)
pi = []
for i in N:

    x = np.random.random(i)*2-1
    y = np.random.random(i)*2-1
    r = np.sqrt(x**2 + y**2)

    povrsina_unutra = len(r[r<1])
    pi.append(4*povrsina_unutra/i)

plt.figure()
plt.plot(N,pi)
plt.xlabel('broj pokusaja')
plt.ylabel('pi')
plt.show()
```

### Izracunati integral funkcije f(x) = x^3 na segmentu [4,19] pomocu Monte Karlo metode.
### ( analiticka vrednost integrala I =  32516.3 )

```python

def funkcija(x):
    return x**3

a1 = 4
a2 = 19

s = (a2-a1)*max(funkcija(a1),funkcija(a2))
N = np.arange(1000,100000,1000)
hit_or_miss = np.zeros(len(N))

for i in range(len(N)):
    n = N[i]
    x = np.random.random(n)*(a2-a1) + a1
    y = np.random.random(n)*max(funkcija(a1),funkcija(a2))

    F = funkcija(x)
    hit = len(y[y<F])
    hit_or_miss[i] = hit/n*s

plt.plot(N,hit_or_miss)
plt.show()
```
