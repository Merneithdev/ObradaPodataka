### Generisati uzorak cija je funkcija raspodele data sa F(x) = cosx + 1/pi

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
velicina_uzorka = 100000
x = np.linspace(0,np.pi,1000)
pdf = (np.cos(x) + 1)/np.pi
cdf = (np.sin(x) + x)/np.pi

y = np.linspace(min(cdf),max(cdf),velicina_uzorka)


###### NJUTN #####
uzorak = np.zeros_like(y)
for i in range(len(y)):
    xx = y[i]
    delta = 1
    while abs(delta)>1e-3:
        f = (np.sin(xx) + xx)/np.pi - y[i]    # f = cdf - 1
        f1 = (np.cos(xx) +1)/np.pi
        delta = f/f1
        xx = xx - delta
    uzorak[i] = xx

    
plt.plot(x,pdf,'--k',label = 'gustina raspodele')
plt.plot(x,cdf,label='kummulativna funkcija raspodele')
plt.hist(uzorak,100,          density=True,          label='generisani uzorak')
plt.legend()
plt.show()
```
